```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_87t = pd.read_csv("/content/drive/MyDrive/PLANT RESEARCH/datasets/processed_species_87t.csv")
```

```python
summary_data = pd.read_excel("/content/drive/MyDrive/PLANT RESEARCH/datasets/2025-01-20-Exported Summary Data_all species.201t.FN.xlsx")
```

```python
summary_data.head(4)
```

```python
#Do the left merge
data_87t = (
    data_87t
    .merge(
        summary_data[['species', 'speciation']],
        how='left',
        left_on='species',
        right_on='species'
    )
)
```

```python
# Calculate the correlation matrix of the normalized data
corr_matrix = data_87t[[ 'Rare (Yes/No)', 'speciation','log.b19', 'log.b18', 'log.b17',
       'log.b16', 'log.b15', 'log.b14', 'log.b13', 'log.b12', 'log.b11',
       'log.b10', 'log.b9', 'log.b8', 'log.b7', 'log.b6', 'log.b5', 'log.b4',
       'log.b3', 'log.b2', 'log.b1']].corr(method = 'pearson')


plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin = -1, vmax = 1)

# Display the heatmap
plt.title('Correlation Between Rarity and Each bio variable_87 species only in Western Australia')
plt.show()
```
