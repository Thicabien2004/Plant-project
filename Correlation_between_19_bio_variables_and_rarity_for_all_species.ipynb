```python
world_clim_2506 = pd.read_csv('/content/drive/MyDrive/PLANT RESEARCH/datasets/Copy of Rham.ploid.midp.WorldClim.2506.csv')
```

```python
#Do the left merge
world_clim_2506 = (
    world_clim_2506
    .merge(
        summary_data[['species', 'Rare (Yes/No)']],
        how='left',
        left_on='species',
        right_on='species'
    )
)
```

```python
world_clim_2506['Rare (Yes/No)'] = world_clim_2506['Rare (Yes/No)'].str.lower().map({'yes': 1, 'no': 0})
```

```python
world_clim_2506['Rare (Yes/No)'].value_counts()
```

```python
# Calculate the correlation matrix of the normalized data
corr_matrix = world_clim_2506[[ 'Rare (Yes/No)','log.b19', 'log.b18', 'log.b17',
       'log.b16', 'log.b15', 'log.b14', 'log.b13', 'log.b12', 'log.b11',
       'log.b10', 'log.b9', 'log.b8', 'log.b7', 'log.b6', 'log.b5', 'log.b4',
       'log.b3', 'log.b2', 'log.b1']].corr(method = 'pearson')


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin = -1, vmax = 1)

# Display the heatmap
plt.title('Correlation Between Rarity and Each bio variable')
plt.show()
```

```python
summary_data['Rare (Yes/No)'] = summary_data['Rare (Yes/No)'].str.lower().map({'yes': 1, 'no': 0})
```

```python
summary_data['Rare (Yes/No)'].value_counts()
```

```python
summary_data_1 = summary_data.copy()
# List of relevant columns
cols = ['Rare (Yes/No)', 'Length (km)', 'AGE (Ma)', 'ploidy', 'Area (km2)',
        'turnover', 'net.div', 'speciation', 'extinction']

# Convert to numeric and coerce errors (e.g., blank strings)
for col in cols:
    summary_data_1[col] = pd.to_numeric(summary_data_1[col], errors='coerce')

# Drop rows with any NaN values in the selected columns
summary_data_clean = summary_data_1.dropna(subset=cols)

# Compute correlation matrix
corr_matrix = summary_data_clean[cols].corr(method='pearson')

# Plot
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Between Rarity and Other Variables (All Species Around Australia)')
plt.show()
```
